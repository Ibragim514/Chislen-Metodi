### Практическая работа № 2

```python
from math import *

def f(x):
    # Уравнение приведено к виду f(x) = 0
    result = 10 - 0.5 * x**2 - 2**(-x)
    return result

# Выбираем интервал, где функция меняет знак
# Например, для x=4: f(4) = 10 - 8 - 0.0625 = 1.9375
# Для x=5: f(5) = 10 - 12.5 - 0.03125 = -2.53125
# Значит, корень находится между 4 и 5
a = 4
b = 5
epsilon = 0.001 

while (b - a) / 2 > epsilon:
    c = (a + b) / 2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    
Ответ = (a + b) / 2
print(f"Корень: {Ответ}")
```

### Практическая работа № 3

```python
from math import *

# Преобразуем уравнение к виду x = ϕ(x)
def f(x):
    return sqrt(20 - 2 * 2**(-x))

x1 = 0
x2 = 4.5  # Начальное приближение (корень находится около 4.47)
count = 1
epselant = 10**(-3)

print(f'0 - {x2}')

while abs(x2 - x1) > epselant:
    x1 = x2
    x2 = f(x1)
    print(f'{count} - {round(x2, 4)}')
    count += 1

print(f'Ответ: {round(x2, 4)}')
```
