```python
# Задание 2.2

from math import *

def f(x):
  y = x*sin(x) - 1
  return y

eps = 1
a = 1
b = 1.2
while eps > 0.0001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  eps = (b - a)/2
Ответ = (a + b)/2
print(round(Ответ,4))
```

```python
# Задание 2.3

from math import *

def f(x):
  y = 1/sin(x)
  return y

x1 = 1
# х2 = х0
x2 = 1.1
c = 1
eps = 10**(-5)
print(f'0 --- {x2}')
while abs(x2 - x1) > eps:
    x1 = x2
    x2 = round(f(x1), 5)
    print(f'{c} --- {x2}')
    c += 1
print(f'Ответ: {x2}')
```

```python
# Задание 2.5

from math import *
def f(x):
  y = x**3 + x**2 - 24*x + 1
  return y
xl = float(input('Введите левый х: ')) # 0
x2 = float(input('Введите правый х: ')) # 1
eps = float(input('Введите эпселант: ')) # 0.000001
d = (xl-x2)/2
n = 0
c = xl
print(f'n = {n} | xl = {xl} | x2 = {x2}')
while abs(d) > eps:
  xl = round(xl - f(xl)/(3*xl**2 + 6*xl - 24), 6)
  x2 = round((c*f(x2) - x2*f(c))/(f(x2) - f(c)), 6)
  y = (xl+x2)/2
  d = (xl-x2)/2
  n += 1
  print(f'n = {n} | xl = {xl} | x2 = {x2}')
x = y
print(f'x = {x}  n = {n}')
```
