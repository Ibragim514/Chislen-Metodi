```python
# Задание 2.2

from math import *

def f(x):
  y = x*sin(x) - 1
  return y

eps = 1
a = 1
b = 1.2
while eps > 0.0001:
  c = (a + b)/2
  if f(a) * f(c) < 0:
    b = c
  else:
    a = c
  eps = (b - a)/2
Ответ = (a + b)/2
print(round(Ответ,4))
```

```python
# Задание 2.3

from math import *

x = 1.0
eps = 1e-5 # 1e-5 = 1*10 в -5 степени
n = 0

while True:
    print(f"{n} --- {x:.6f}")
    sled_x = x - 0.5 * (x * sin(x) - 1)
    if abs(sled_x - x) < eps:
        break
    x = sled_x
    n += 1
```

```python
# Задание 2.5

from math import *
def f(x):
  y = x**3 + x**2 - 24*x + 1
  return y
xl = float(input('Введите левый х: '))
xr = float(input('Введите правый х: '))
eps = float(input('Введите эпселант: '))
d = (xl-xr)/2
n = 0
c = xl
print(f'n = {n} | xl = {xl} | xr = {xr}')
while abs(d) > eps:
  xl = round(xl - f(xl)/(3*xl**2 + 6*xl - 24), 6)
  xr = round((c*f(xr) - xr*f(c))/(f(xr) - f(c)), 6)
  y = (xl+xr)/2
  d = (xl-xr)/2
  n += 1
  print(f'n = {n} | xl = {xl} | xr = {xr}')
x = y
print(f'x = {x}  n = {n}')
```
